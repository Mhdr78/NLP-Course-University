```python
from whatlies import Embedding, EmbeddingSet
import csv
import numpy as np
from collections import Counter
```

```python
import pickle

def load_model(path):
    with open(path, 'rb') as file:
        data = pickle.load(file)
    return data

data_m = load_model('/content/saved_weights_m_pytorch.pkl')
embeddings = data_m['embeddings']
metadata = list(embeddings.keys())
vectors = list(embeddings.values())
```

```python
len(metadata)
```

```python
embeddings = {w:Embedding(w,vectors[i]) for i, w in enumerate(metadata) }
```

```python
# words = ['تشريف', 'صراط', 'للقرآن']
emb_set = EmbeddingSet(*[embeddings[word] for word in metadata])
emb_set
```
